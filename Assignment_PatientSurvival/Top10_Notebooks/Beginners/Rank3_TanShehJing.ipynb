## Handling null values

### Patients with null Treated_with_drugs

```python
data[data.Treated_with_drugs.isna()]
```

For records with null values in column 'Treated_with_drugs', the patients both have previous condition of Z, both came from RURAL area, without revealing their smoking condition and survived after one year. Therefore, we will look into pther patients with these conditions to figure out how could we replace the null value. 

```python
data[data.Z == 1.0]
```

```python
data.Z.value_counts()
```

We found out that patients with previous condition of Z are having nulls in the column 'treated_with_drugs', therefore, this column could not give much information to replace the nulls.

```python
a = data[(data.Survived_1_year == 1) & (data.Patient_Rural_Urban == 'RURAL') & (data.Patient_mental_condition == 'Stable')]
a
```

```python
a.Treated_with_drugs.value_counts()
```

We will fill the nulls in 'Treated_with_drugs' using the mode of treated drugs on patients who came from rural area and survived after 1 year. (The drug will be DX6)

```python
data.Treated_with_drugs = data.Treated_with_drugs.fillna('DX6')
```

Among these patients, most of the survived patients have 1 previous condition and most of it are condition A. Therefore, we will replace the null value with 1 in A and 0 for others. While the column Number_of_prev_cond are calculated based on number of 1s in column A,B,C,D,E,F,Z.

```python
# fill nulls in column A
data.A = data.A.fillna(1.0)
```

```python
# fill nulls for other columns
data.B = data.B.fillna(0.0)
data.C = data.C.fillna(0.0)
data.D = data.D.fillna(0.0)
data.E = data.E.fillna(0.0)
data.F = data.F.fillna(0.0)
data.Z = data.Z.fillna(0.0)
```

```python
data.Number_of_prev_cond = data.Number_of_prev_cond.fillna(1.0)
```

### Convert binary categorical column to numeric

```python
def smoke_to_numeric(x):
    if x=='NO':
        return 0
    elif x=='YES':
        return 1
    else:
        return 2
```

```python
data.Patient_Smoker = data.Patient_Smoker.apply(smoke_to_numeric)
```

```python
def area_to_numeric(x):
    if x=='URBAN':
        return 1
    elif x=='RURAL':
        return 0
```

```python
data.Patient_Rural_Urban = data.Patient_Rural_Urban.apply(area_to_numeric)
```

```python
# drop the column 'Patient_mental_condition' as it has only one value in this column, therefore it will not have any effect on the dataset

data = data.drop(columns=['Patient_mental_condition'])
```

```python
# convert columns A-Z to int instead of float
data.A = data.A.astype(int)
data.B = data.B.astype(int)
data.C = data.C.astype(int)
data.D = data.D.astype(int)
data.E = data.E.astype(int)
data.F = data.F.astype(int)
data.Z = data.Z.astype(int)
```

```python
data.insert(3, "Drug_DX1", True) 
data.insert(4, "Drug_DX2", True)
data.insert(5, "Drug_DX3", True)
data.insert(6, "Drug_DX4", True)
data.insert(7, "Drug_DX5", True)
data.insert(8, "Drug_DX6", True)
```

```python
data
```

```python
data.Drug_DX1 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX1' in x.split() else 0)
data.Drug_DX2 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX2' in x.split() else 0)
data.Drug_DX3 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX3' in x.split() else 0)
data.Drug_DX4 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX4' in x.split() else 0)
data.Drug_DX5 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX5' in x.split() else 0)
data.Drug_DX6 = data.Treated_with_drugs.apply(lambda x: 1 if 'DX6' in x.split() else 0)
```

```python
data = data.drop(columns=['Treated_with_drugs'])
```

## Baseline model - Random Forest Classifier

```python
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score

from sklearn.feature_selection import SelectFromModel
```

```python
# split the data into X and y
X = data.drop(columns=['Survived_1_year'])
y = data['Survived_1_year']

# Split randomly into 80% train data and 20% test data
X_train, X_Test, y_train, y_Test = train_test_split(X, y, test_size = 0.2, random_state = 57)
```

```python
rfc = RandomForestClassifier(random_state=1, n_jobs=-1, criterion='entropy')
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_Test)
f1_score(y_Test, y_pred)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(rfc.feature_importances_, index=X_train.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
# Create a selector object that will use the random forest classifier to identify
# It will select the features based on the importance score
rfc_sfm = SelectFromModel(rfc)

#Alternatively if you want to set a specific threshold and select the features you may the following code that is commented
# features that have an importance of more than 0.13 ( all top 4 features lie above this)
# rf_sfm = SelectFromModel(rf_clf, threshold=0.13)

# Train the selector
rfc_sfm = rfc_sfm.fit(X_train, y_train)
```

```python
# Transform the data to create a new dataset containing only the most important features
# Note: We have to apply the transform to both the training X and test X data.
X_important_train = rfc_sfm.transform(X_train)
X_important_test = rfc_sfm.transform(X_Test)
```

```python
# Create a new random forest classifier for the most important features
clf_important = RandomForestClassifier(random_state=1, n_jobs=-1, criterion='entropy')

# Train the new classifier on the new dataset containing the most important features
clf_important = clf_important.fit(X_important_train, y_train)
```

```python
y_pred = rfc.predict(X_Test)
rf_ac = accuracy_score(y_Test, y_pred)
rf_fscore = f1_score(y_Test ,y_pred)

# View The Accuracy Of Our Full Feature (8 Features) Model
print("Baseline Model Accuracy:", rf_ac)
print("Baseline Model F1 Score:", rf_fscore)
```

```python
# Apply The Full Featured Classifier To The Test Data
y_important_pred = clf_important.predict(X_important_test)

# View The Accuracy Of Our Limited Feature (4 Features) Model
rf_imp_ac = accuracy_score(y_Test, y_important_pred)
rf_imp_fscore = f1_score(y_Test, y_important_pred)

print("Limited Features Model Accuracy:", rf_ac)
print("Limited Features Model F1 Score:", rf_fscore)
```

```python
# Evaluating the model
predictions = selection_model.predict(select_X_test)      # predict for selected features of X_test
xgb_accuracy = accuracy_score(y_Test, predictions)        # Calculating accuracy score on test data
xgb_fscore = f1_score(y_Test , predictions)               # Calculating F1 score on test data


# View The Accuracy Of Our Limited Feature (4 Features) Model
print("Limited Features XGBoost Model Accuracy:", xgb_accuracy)
print("Limited Features XGBoost Model F1 Score:", xgb_fscore)
```
